### Overview of the data - Univariate analysis

```python
plt.figure(figsize=(10,4))

ax = sns.countplot(x='MSZoning', data=df1, order=df1['MSZoning'].value_counts(ascending=False).index)
values = df1['MSZoning'].value_counts(ascending = False).values
ax.bar_label(container = ax.containers[0], labels = values)
plt.title('Number of MSZoning');
```

##### **Observations**
- We can observe that Residential High Density (RL) has the highest number of all the zoning classification followed by the Residential Medium Density (RM) then Commercial (C). 

```python
fig, ax = plt.subplots(figsize = (3,3))
count = Counter(df1["MSZoning"])

ax.pie(count.values(), labels = count.keys(), autopct = lambda p:f'{p:.2f}%');
```

#### Observations
- We can observe from the above visualization that 78.82% of the houses from the dataset fall within RL

```python
plt.figure(figsize=(10,4))

ax = sns.countplot(x='SaleCondition', data=df1, order=df1['SaleCondition'].value_counts(ascending=False).index)
values = df1['SaleCondition'].value_counts(ascending = False).values
ax.bar_label(container = ax.containers[0], labels = values)
plt.title('Number of SaleCondition');
```

##### **Observations**
- In terms of sale condition, Normal has the most count followed by Partial and then Abnormal, then least are Family, Alloca and AdjLand

```python
plt.figure(figsize=(10,4))

ax = sns.countplot(x='SaleType', data=df1, order=df1['SaleType'].value_counts(ascending=False).index)
ax.bar_label(ax.containers[0])             # Add labels to bars
plt.title('Number of SaleType');
```
